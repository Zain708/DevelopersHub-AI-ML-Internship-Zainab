# Task 1: Exploring and Visualizing a Simple Dataset

**Problem Statement:** Before applying any machine learning models, an AI/ML engineer must understand the underlying structure, relationships, and anomalies within their data.

**Goal:** The objective of this notebook is to load the Iris dataset, perform statistical inspection, and build visualizations (scatter plots, histograms, and box plots) to uncover patterns, feature distributions, and potential outliers among the different flower species.

```python
import pandas as pd # for the data manipulation like putting the data in tables (Excel Sheet)
import seaborn as sns # for building plots and also it has iris dataset ready for use
import matplotlib.pyplot as plt # for builing plots and seaborn use this as it's foundation or engine
df = sns.load_dataset('iris') # loading the dataset using seaborn's built-in function sns.load_dataset('name of dataset') in variable df (DataFrame)
print(df.head(5)) # printing the first 5 rows of dataset using pandas' .head() function
```

```python
# Using Pandas built-in tools to inspect the dataset:
# - Check total size and missing data
# - Calculate statistical averages (like petal length)

# - A attribute is a fact (it is something), It does not use parentheses because it just looks up a piece of information that is already stored
dataset_shape = df.shape # tells you the dimensions of the dataset as (rows, columns)
dataset_columns = df.columns # lists all the column names
print(dataset_shape)
print(dataset_columns)

# - A function is an action (it does something), It uses parentheses () because it has to run code to calculate a result
print(df.info()) # This function (needs parentheses) gives a concise summary of the DataFrame, including the data types of each column and if there are any missing (null) values
print(df.describe()) # # This function calculates summary statistics (like mean, minimum, maximum, and percentiles) for all the numerical columns
```

```python
sns.scatterplot(data = df, x = 'sepal_length', y = 'sepal_width', hue = 'species' )
plt.show() # plot displayed by matplotlib
```

```python
sns.histplot(data = df,x = 'petal_length', kde = True)
plt.show()
```

```python
sns.boxplot(data = df, x ='species', y = 'sepal_length')
plt.show()
```

```python
sns.boxplot(data = df, x ='species', y = 'sepal_width')
plt.show()
```

# Conclusion & Final Insights

Based on our Exploratory Data Analysis (EDA) of the Iris dataset, we discovered the following key insights:

1. **Data Quality:** The dataset contains 150 samples with 5 columns. It is perfectly clean with absolutely zero missing (null) values.
2. **Feature Distributions:** The histogram of `petal_length` revealed a bimodal distribution (two distinct peaks).
3. **Species Separability:** Scatter plots showed that the `Setosa` species is highly distinct and isolated from the other two species, making it very easy for an AI to classify. `Versicolor` and `Virginica` have some overlap.
4. **Outlier Detection:** Box plots revealed that while `sepal_length` had no anomalies, the `Setosa` species contains a few specific outliers in its `sepal_width` measurements.
