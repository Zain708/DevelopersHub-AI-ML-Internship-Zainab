# Task 4: Health Query Chatbot

**Problem Statement:** Accessing quick, basic medical advice can sometimes be difficult. While an AI cannot replace a doctor, a simple automated system can help triage common symptoms and provide immediate, standard care advice.

**Goal:** Build a rule-based Natural Language chatbot using Python. The bot will take user input, normalize the text, scan for specific symptom keywords (like fever, headache, or cough), and return pre-defined health advice using conditional logic and dictionary mapping.

```python
def health_chatbot(user_input):
  # 1. Convert the user's text to lowercase so our bot doesn't get confused by capital letters
  user_input = user_input.lower()
  # 2. The Bot's Knowledge Base (Dictionary mapping keywords to advice)
  knowledge_base = {
      "fever" : "It sounds like you have a fever. Please rest, drink plenty of fluids, and take a fever reducer. If it lasts over 3 days, see a doctor.",
      "headache": "For a headache, try drinking water, resting in a quiet dark room, and taking over-the-counter pain relief. Limit screen time!",
      "cough": "For a cough, try honey with warm water. If it's a dry cough, throat lozenges might help. See a doctor if you have difficulty breathing.",
      "stomach": "For an upset stomach, stick to bland foods like bananas, rice, applesauce, and toast (the BRAT diet). Stay hydrated!" ,
      "cold": "For a cold, get plenty of rest, stay hydrated with warm fluids, and use a saline nasal spray for congestion. Antibiotics will not help a cold virus.",
      "sore throat": "To soothe a sore throat, gargle with warm salt water, drink warm tea with honey, and try throat lozenges. Avoid crunchy or highly acidic foods.",
      "fatigue": "If you are feeling constantly tired, ensure you are getting 7-9 hours of sleep, eating balanced meals, and staying hydrated. Persistent fatigue should be checked by a doctor.",
      "nausea": "For nausea, sip on ginger ale or peppermint tea, sit upright after eating, and avoid strong cooking smells. Try eating small, frequent meals instead of large ones.",
      "diarrhea": "With diarrhea, the main goal is preventing dehydration. Drink electrolyte solutions or clear broths. Avoid dairy, greasy foods, and caffeine until you feel better.",
      "allergy": "For seasonal allergies, limit outdoor time when pollen counts are high, keep windows closed, and consider over-the-counter antihistamines. Wash your clothes after being outside.",
      "muscle ache": "For body or muscle aches, try a warm bath, gentle stretching, and over-the-counter pain relievers. Rest the affected muscles and avoid heavy lifting.",
      "dizziness": "If you feel dizzy, sit or lie down immediately to prevent falling. Drink water, avoid sudden movements, and do not drive. See a doctor if it is severe or frequent.",
      "heartburn": "For heartburn or acid reflux, avoid lying down for at least 3 hours after eating. Steer clear of spicy, fatty, or mint-flavored foods, and eat smaller portions.",
      "insomnia": "To help with sleep trouble, keep your bedroom dark and cool, avoid screens 1 hour before bed, and skip caffeine in the afternoon. Try to go to bed at the exact same time daily."
  }
  # 3. Scan the user's input for any of our keywords
  for keyword, advice in knowledge_base.items():
    if keyword in user_input:
      return advice  # If we find a match, immediately give the advice!

  # 4. Fallback response if the bot doesn't recognize the symptoms
  return "I am a simple health bot and don't recognize those symptoms. Please consult a real healthcare professional!"
```

```python
# 1. Ask the user for their symptoms
user_input = input('Enter your problem here: ')

# 2. Print what they typed
print(f"You said: {user_input}")

# 3. Get the bot's response and print it to the screen!
bot_response = health_chatbot(user_input)
print(f"Bot says: {bot_response}")
```

# Conclusion & Final Insights

By building this rule-based chatbot, we demonstrated how foundational Python logic can simulate a conversational agent:
1. **Text Normalization:** Converting user input to lowercase ensures that our bot doesn't fail due to arbitrary capitalization.
2. **Keyword Mapping:** Using a dictionary allowed us to efficiently link symptoms (keys) to medical advice (values).
3. **Control Flow:** The `for` loop and `in` keyword allowed the bot to scan entire sentences, meaning the user doesn't have to type just a single word to get an answer.
4. **Limitations:** Because this is rule-based and not a Machine Learning Natural Language Processing (NLP) model, it lacks contextual understanding and will fail if the user mispells a word or uses a synonym (e.g., "migraine" instead of "headache").

# With NLP(Natural Language Processing) Machine learning model using Scikit-Learn

# Task 4: Health Query Intent Classification (NLP)

**Problem Statement:** Traditional rule-based chatbots fail when users type unexpected words or synonyms, making them brittle and frustrating to use for healthcare queries.

**Goal:** Build a Natural Language Processing (NLP) Machine Learning model using Scikit-Learn. We will use TF-IDF vectorization to convert user text into numerical features, and train a Logistic Regression model to classify the underlying "intent" of the user's symptoms (e.g., fever, headache, cold) to provide the correct medical advice.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# 1. Create a tiny training dataset (Text and the "Intent" or label)
data = {
    'text': [
        "I have a terrible fever", "My temperature is really high",
        "My head is pounding", "I have a severe migraine",
        "I can't stop coughing", "My throat hurts and I am hacking",
        "My tummy hurts", "I have awful stomach cramps",
        "My nose is running", "I keep sneezing and have a runny nose"
    ],
    'intent': [
        "fever", "fever",
        "headache", "headache",
        "cough", "cough",
        "stomach", "stomach",
        "cold", "cold"
    ]
}
df_chat = pd.DataFrame(data)

# 2. Build the NLP Pipeline: Text goes in -> turns to numbers -> ML Model predicts
# We are using LogisticRegression as our classification model!
nlp_model = make_pipeline(TfidfVectorizer(), LogisticRegression())

# 3. Train the model on our dataset!
nlp_model.fit(df_chat['text'], df_chat['intent'])

# 4. The Bot's Knowledge Base for giving advice
advice_dictionary = {
    "fever": "Rest, drink fluids, and take a fever reducer.",
    "headache": "Drink water, rest in a dark room, and limit screen time.",
    "cough": "Try honey with warm water or throat lozenges.",
    "stomach": "Stick to bland foods like rice and toast. Stay hydrated!",
    "cold": "It sounds like a cold or allergies. Try taking an antihistamine and blow your nose often!"
}

# 5. Let's test the model!
# Notice we are using a completely new sentence it was NOT trained on!
user_input = ["My skull is absolutely aching today"]

# Predict the intent using the ML model
predicted_intent = nlp_model.predict(user_input)[0]

print(f"User said: {user_input[0]}")
print(f"Model classified the intent as: {predicted_intent}")
print(f"Bot Advice: {advice_dictionary[predicted_intent]}")
```

# Conclusion & Final Insights

By upgrading from a rule-based system to a Machine Learning NLP model, we learned the following:
1. **Text Vectorization:** We used `TfidfVectorizer` to convert raw text into a mathematical matrix that the AI could process.
2. **Intent Classification:** Instead of looking for exact keyword matches, the `LogisticRegression` model learned to predict the category (intent) of the entire sentence.
3. **Model Limitations:** Through stress-testing, we discovered that TF-IDF relies heavily on exact vocabulary from its training data. When given completely unseen words (like "skull" instead of "head"), it was forced to make mathematical guesses based on grammar ("my" and "is").
4. **The Solution:** We successfully retrained the model by adding more diverse data (e.g., a "cold" intent with runny nose symptoms) to expand its vocabulary and accuracy. True semantic understanding would require more advanced Deep Learning Word Embeddings.
