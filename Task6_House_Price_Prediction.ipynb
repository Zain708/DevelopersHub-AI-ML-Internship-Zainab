# Task 6: House Price Prediction (Linear Regression)

**Problem Statement:** Real estate pricing depends heavily on the size of the property, but calculating exact estimates manually can be tedious.

**Goal:** The objective of this notebook is to build a foundational Machine Learning model using Scikit-Learn. We will create a dataset of house sizes and prices, train a Linear Regression model to find the mathematical relationship between them, and use the trained model to predict the price of a completely unseen house (2000 sqft).

```python
import pandas as pd # For creating data tables (DataFrames) to manage house features like square footage and prices
import matplotlib.pyplot as plt # For plotting graphs to visualize our AI's prediction line against actual data
from sklearn.linear_model import LinearRegression # From Scikit-Learn (the ultimate ML library), import the algorithm to build our predictive model
```

```python
# Raw data of houses: square footage vs price (in $1,000s) in dictionary
house_data = {
    'sqft': [1000, 1500, 1800, 2400, 3000],
    'price': [200, 290, 350, 470, 590]
}
```

```python
df_houses = pd.DataFrame(house_data) # Converting the raw data which is in dictionary into a DataFrame (proper table form like in columns and rows) using pandas' pd.DataFrame() function and in the variable df_houses
print(df_houses.head()) # Using the pandas' function .head() to print the table
```

```python
# 1. Fix the DataFrame creation by adding house_data inside the parentheses
df_houses = pd.DataFrame(house_data)

# 2. Extract the features (Inputs) - Notice the double brackets!
X = df_houses[['sqft']]

# 3. Extract the target (Outputs) - Notice the single brackets!
y = df_houses['price']

# Let's print X to make sure it looks like a clean feature column
print("--- Features (X) ---")
print(X)
```

```python
# 1. Create an empty Linear Regression model instance
model =  LinearRegression()

# 2. Train the model using our features (X) and target (y)
model.fit(X,y)

# 3. Let's test the model! Let's ask it to predict the price of a 2000 sqft house.
# Note: We pass [[2000]] in double brackets because the input must be 2D!
predicted_price = model.predict([[2000]])

print(f"The model predicts a 2000 sqft house will cost: ${predicted_price[0]:.2f} thousand dollars!")
```

```python
plt.scatter(X, y, color ='blue', label ='Actual Houses')
plt.plot(X, model.predict(X), color = 'red', label ='Predicted Houses')
# We use the 2000 sqft and the predicted_price[0] we calculated earlier
plt.scatter(2000, predicted_price[0], color='green', marker='x', s=30, label='Predicted 2000 sqft House')
plt.title('House Price Prediction: Sqft vs. Price')
plt.xlabel('Square Footage')
plt.ylabel('Price in ($1,000s)')
plt.legend()
plt.show()
```
